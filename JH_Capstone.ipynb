```python
import pandas as pd
from matplotlib import pyplot
from pandas.plotting import scatter_matrix
from sklearn import linear_model, metrics, model_selection, svm
from pandas.plotting import scatter_matrix

import matplotlib.pyplot as plt

df = pd.read_csv('/content/sample_data/diabetes_Data.csv')
mylog_model = linear_model.LogisticRegression(max_iter=1000)
mysvm_model = svm.SVC(max_iter=1000)
y = df.values[:,8]
X = df.values[:,0:8]
X_train, X_test, y_train, y_test = model_selection.train_test_split(X,y,test_size=0.3)

mylog_model.fit(X_train, y_train)
mysvm_model.fit(X_train, y_train)

y_pred_log = mylog_model.predict(X_test)
y_pred_svm = mysvm_model.predict(X_test)


print(metrics.accuracy_score(y_test, y_pred_log))
print(metrics.accuracy_score(y_test, y_pred_svm))


value = input("Enter a value for: Number of pregnancies(as a whole number), Glucose, Blood Pressure, Skin Thickness, Insulin, BMI, Diabetes Pedigree Function, and Age. (Seperated with a space):\n")
valueList = list(value.split(" "))
print(mysvm_model.predict([valueList]))
```

```python
import pandas as pd
from matplotlib import pyplot
from pandas.plotting import scatter_matrix
from sklearn import linear_model, metrics, model_selection, svm
from sklearn.metrics import ConfusionMatrixDisplay
df = pd.read_csv('diabetes.csv')


import matplotlib.pyplot as plt


mylog_model = linear_model.LogisticRegression(max_iter=1000)
mysvm_model = svm.SVC(max_iter=1000)
y = df.values[:,8]
X = df.values[:,0:8]
X_train, X_test, y_train, y_test = model_selection.train_test_split(X,y,test_size=0.3)

mylog_model.fit(X_train, y_train)
mysvm_model.fit(X_train, y_train)

y_pred_log = mylog_model.predict(X_test)
y_pred_svm = mysvm_model.predict(X_test)


print(metrics.accuracy_score(y_test, y_pred_log))
print(metrics.accuracy_score(y_test, y_pred_svm))



ConfusionMatrixDisplay.from_estimator(mysvm_model, X_test, y_test);
```
